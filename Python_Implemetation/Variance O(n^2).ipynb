```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
B = 500
n_lst = [10, 50, 100, 500, 1000, 5000, 10000]
n_lst.reverse()
mean_var = []
np.random.seed(123)
x_qry = np.linspace(0, 1, 100)
Y_true1 = (np.cos(2*np.pi*x_qry))**2 + np.log(4/3 + x_qry)

plt.rcParams.update({'font.size': 12})  # Change the font sizes of ouput figures
plt.figure(figsize=(8,6))
# plt.plot(x_qry, Y_true1, color='red', label='True curve')

for i in range(len(n_lst)):
    X_1 = np.linspace(0, 1, n_lst[i])
    m_1 = (np.cos(2*np.pi*X_1))**2 + np.log(4/3 + X_1)
    Y_hat = np.zeros((B, n_lst[i]-1))
    for b in range(B):
        eps1 = np.random.normal(0, 0.1, n_lst[i])
        Y_1 = m_1 + eps1
        Y_hat[b,:] = (Y_1[1:] - Y_1[:(n_lst[i]-1)])/(X_1[1:] - X_1[:(n_lst[i]-1)])
    
    X_cen = (X_1[1:] + X_1[:(n_lst[i]-1)])/2
    plt.scatter(X_cen, np.mean(Y_hat, axis=0), color='blue')
    plt.errorbar(X_cen, np.mean(Y_hat, axis=0), yerr=np.var(Y_hat, axis=0), 
                 capsize=5, elinewidth=1, label='Diff Quo Est (n='+str(n_lst[i])+')')
    
    mean_var.append(np.mean(np.var(Y_hat, axis=0)))
    
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
B = 500
n_lst = [10, 50, 100, 200, 500, 800, 1000, 2000, 5000, 8000, 10000]
mean_var = []
np.random.seed(123)

for i in range(len(n_lst)):
    X_1 = np.linspace(0, 1, n_lst[i])
    m_1 = (np.cos(2*np.pi*X_1))**2 + np.log(4/3 + X_1)
    Y_hat = np.zeros((B, n_lst[i]-1))
    for b in range(B):
        eps1 = np.random.normal(0, 0.1, n_lst[i])
        Y_1 = m_1 + eps1
        Y_hat[b,:] = (Y_1[1:] - Y_1[:(n_lst[i]-1)])/(X_1[1:] - X_1[:(n_lst[i]-1)])
    
    mean_var.append(np.mean(np.var(Y_hat, axis=0)))
```

```python
x = np.linspace(1, max(n_lst), 100)

plt.figure(figsize=(8,6))
plt.scatter(n_lst, mean_var)
plt.plot(x, 0.02*(x-1)**2, linestyle='dashed', color='red')
plt.xlabel('Sample size n')
plt.ylabel('Mean variance of the difference quotients across all observations')
plt.title('Equispaced design')
plt.show()
```

```python
B = 1000
n_lst = [10, 50, 100, 200, 500, 800, 1000, 2000, 4000, 5000, 6000, 7000, 8000, 10000]
mean_var_ran = []
np.random.seed(123)

for i in range(len(n_lst)):
    Y_hat = np.zeros((B, n_lst[i]-1))
    X_r = np.random.rand(n_lst[i])
    X_r = np.sort(X_r)
    m_r = (np.cos(2*np.pi*X_r))**2 + np.log(4/3 + X_r)
    for b in range(B):
        eps1 = np.random.normal(0, 0.1, n_lst[i])
        Y_r = m_r + eps1
        Y_hat[b,:] = (Y_r[1:] - Y_r[:(n_lst[i]-1)])/(X_r[1:] - X_r[:(n_lst[i]-1)])
    
    mean_var_ran.append(np.mean(np.var(Y_hat, axis=0)))
```

```python
x = np.linspace(1, max(n_lst), 100)

plt.figure(figsize=(8,6))
plt.scatter(n_lst, mean_var_ran)
plt.plot(x, (x-1)**2, linestyle='dashed', color='red')
plt.plot(x, (x-1)**3, linestyle='dashed', color='blue')
plt.xlabel('Sample size n')
plt.ylabel('Mean variance of the difference quotients across all observations')
plt.title('Random design')
plt.show()
```
